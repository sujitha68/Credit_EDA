**Data Description**
Dataset contain 28 columns

**1. ID:** A unique identifier for each record or transaction.
Type: int64

**2. Customer_ID:** A unique identifier for each customer. Multiple rows with the same Customer_ID represent different months for the same customer.
Type: int64

**3. Month:** The month of the recorded transaction or data point, likely represented as a numeric value.
Type: int64

**4. Name:** The name of the customer.
Type: object

***5. Age: ***The age of the customer.
Type: float64

**6. SSN**:The Social Security Number of the customer. Typically sensitive information that should be masked or encrypted.
Type: float64

**7. Occupation:**The occupation or job title of the customer.
Type: object

**8. Annual_Income:**The annual income of the customer, in monetary units.
Type: float64

***9. Monthly_Inhand_Salary: ***The monthly salary received by the customer after deductions (i.e., the take-home pay).
Type: float64

**10. Num_Bank_Accounts:** The number of bank accounts held by the customer.
Type: float64

**11. Num_Credit_Card**:The number of credit cards owned by the customer.
Type: float64

**12. Interest_Rate:** The interest rate applicable to the customer’s loans or credit accounts.
Type: float64

**13. Num_of_Loan: **The number of loans the customer has taken.
Type: float64

**14. Type_of_Loan:**The types of loans the customer has, listed as a comma-separated string. Examples include Auto Loan, Credit-Builder Loan, Personal Loan, and Home Equity Loan.
Type: object

**15. Delay_from_due_date:** The number of days the customer has delayed payments beyond the due date.
Type: float64

**16. Num_of_Delayed_Payment:**The number of payments the customer has delayed.
Type: float64

**17. Changed_Credit_Limit:**A numeric value representing changes to the customer’s credit limit, potentially indicating the percentage of increase or decrease.
Type: float64

**18. Num_Credit_Inquiries:**The number of times the customer’s credit report has been accessed for inquiries, typically by lenders.
Type: float64

**19. Credit_Mix:** The mix of credit types (e.g., Good, Bad, etc.), indicating the diversity of the customer’s credit portfolio.
Type: object

**20. Outstanding_Debt:** The total amount of outstanding debt the customer has.
Type: float64

***21. Credit_Utilization_Ratio: ***The ratio of the customer’s total credit balance to their credit limit, representing how much credit they are using relative to their credit limit.
Type: float64

**22. Credit_History_Age:**The length of the customer's credit history, often measured in years.
Type: float64

**23. Payment_of_Min_Amount:** Indicates whether the customer pays the minimum amount due on their credit accounts (e.g., Yes, No).
Type: object

**24. Total_EMI_per_month:**The total amount of Equated Monthly Installments (EMI) the customer pays each month for their loans.
Type: float64

**25. Amount_invested_monthly:** The amount of money the customer invests each month.
Type: float64

**26. Payment_Behaviour:** A categorical representation of the customer's payment behavior (e.g., Prompt, Delayed).
Type: object

**27. Monthly_Balance:**The balance remaining in the customer's accounts at the end of each month after all transactions and payments.
Type: float64

**28. Credit_Score:**The credit score category of the customer, usually represented as Good, Bad, or Excellent, indicating their creditworthiness.
Type: object

Outlier Handling

```python
#  visualizes outliers for a specific column (col) in the dataset using boxplot
def outlier_visualizer(col, vert=True, grid=True):
    df1[col].plot(kind='box', vert=vert)
    plt.title('{} - Boxplot'.format(col))
    plt.ylabel('Values')
    plt.grid(grid)
    plt.show()

# identifies outliers for a specific column (col) using the Interquartile Range (IQR) method
def outlier_finder(col):
    q1 = np.percentile(df1[col], 25)
    q3 = np.percentile(df1[col], 75)
    iqr = q3 - q1
    left_lim = q1 - iqr*1.5
    right_lim = q3 + iqr*1.5
    outliers = {'index':[], col:[]}

    for i,c in enumerate(df1[col]):
        if c < left_lim:
            outliers['index'].append(df1[col].index[i])
            outliers[col].append(c)
        elif c > right_lim:
            outliers['index'].append(df1[col].index[i])
            outliers[col].append(c)

    outliers_df = pd.DataFrame(outliers)
    outliers_df.set_index('index', inplace=True)

    print('{} - % of outliers: {}%'.format(col, round(outliers_df.shape[0]/df1.shape[0]*100,2)))
    return outliers_df
```

# *Encoding*


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Define the columns to be label encoded
cols = ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score']

# Apply label encoding to each column
for col in cols:
    df1[col] = label_encoder.fit_transform(df1[col])

df1.head()
```

## Remove Highly Correlated Variables

```python
# Compute the correlation matrix
correlation_matrix = df1.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Define the correlation threshold
correlation_threshold = 0.8

# Filter the correlation matrix to include only highly correlated pairs
highly_correlated = correlation_matrix[(correlation_matrix > correlation_threshold) & (correlation_matrix < 1)]

# Print highly correlated data
print("Highly Correlated Data:")
for col in highly_correlated:
    correlated_vars = highly_correlated[col][highly_correlated[col].notnull()]
    for idx, val in correlated_vars.items():
      print(f"{col} and {idx}: {val:.2f}")
```
